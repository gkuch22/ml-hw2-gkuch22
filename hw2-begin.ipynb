```python
X = df_transaction.copy()
X.shape
```

```python
drop_cols = []
threshold = 0.85
for col in X.columns:
    value_counts = X[col].value_counts(normalize=True)
    
    if value_counts.iloc[0] > threshold:
        drop_cols.append(col)

len(drop_cols)
```

```python
X = X.drop(columns=drop_cols)
X.shape
```

```python
threshold = 0.8
na_cols = [col for col in X.columns if X[col].isna().mean() > threshold]
len(na_cols)
```

```python
X = X.drop(columns=na_cols)
X.shape
```

```python
X
```

```python
for col in X.columns:
    nan = X[col].isna().mean()
    print(f"{col}: {nan*100:.2f} nan")
```

```python
for col in X.columns:
    zeros = (X[col] == 0).mean()
    if zeros <= 0.7:
        continue
    print(f"{col}: {zeros*100:.2f} zeros")
```
